```python
#digit recognisation using MNIST dataset
from torchvision import datasets, transforms
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,), (0.5,)),])
trainset = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
```

```python
for image, label in trainloader:
    dataiter = iter(trainloader)
    images, labels = dataiter.next()
    print(type(images))
    print(images.shape)
    print(labels.shape)
    print("------------------")
```

```python
def activation(x):
    return (1/(1+torch.exp(-x)))
```

```python
inputs=images.view(images.shape[0],784)
```

```python
inp=inputs.shape[1]
hiddens=256
outputs=10
weights_1=torch.randn(inp,hiddens)
weights_2=torch.randn(hiddens,outputs)
bias_1=torch.randn(1,hiddens)
bias_2=torch.randn(1,outputs)
```

```python
h=activation(torch.mm(inputs,weights_1)+bias_1)
y=(torch.mm(h,weights_2)+bias_2)
```

```python
print(y)
```

```python
def softmax(y):
    return torch.exp(y)/(torch.sum(torch.exp(y),dim=1)).view(64,1)
```

```python
probabilities=softmax(y)
```

```python
#Buildings Network with Pytorch
from torch import nn
```

```python
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden=nn.Linear(784,256)
        self.output=nn.Linear(256,10)
        self.sigmoid=nn.Sigmoid()
        self.softmax=nn.Softmax(dim=1)
    
    def forward(self,x):
        x=self.hidden(x)
        x=self.sigmoid(x)
        x=self.output(x)
        x=self.softmax(x)
        return x
```

```python
model=Network()
model
```

```python
import torch.nn.functional as F

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(784,256)
        self.output = nn.Linear(256, 10)
        
    def forward(self, x):
        x = F.sigmoid(self.hidden(x))
        x = F.softmax(self.output(x), dim=1)
        return x
```

```python
#Building NN architecture considering 2 hidden layers

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden1 = nn.Linear(784,126)
        self.hidden2 = nn.Linear(126,64)
        self.output = nn.Linear(64, 10)
        
    def forward(self, x):
        x = F.relu(self.hidden1(x))
        x = F.relu(self.hidden2(x))
        x = F.softmax(self.output(x), dim=1)
        return x
```

```python
model=Network()
model
```
